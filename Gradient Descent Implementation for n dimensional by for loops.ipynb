```python
def step_grad(data,alpha,m,c):
    # new m=m-aplha*slope
    slope_m=0;
    slope_c=0;
    for j in range(0,len(data)):
        i=data[j];
        #print(i[1])
        slope_m+=(i[1]-m*i[0]-c)*(-1*i[0]);
        slope_c+=(i[1]-m*i[0]-c)*(-1);
    slope_m=slope_m*(2/len(data));
    slope_c=slope_c*(2/len(data));
    print(slope_m,slope_c,end="");
    m=m-alpha*slope_m;
    c=c-alpha*slope_c;
    return m,c;

def cost(data,m,c):
    total_cost=0;
    M=len(data);
    for i in range(M):
        x=data[i,0];
        y=data[i,1];
        total_cost+=(1/M)*(y-m*x-c)**2;
    return total_cost;
    
        
```

```python
def grad(data,alpha,num_iterations):
    m=0;
    c=0;
    for i in range(0,num_iterations):
        m,c=step_grad(data,alpha,m,c)
        #print(m,c);
        print(cost(data,m,c));
    return m,c;
```

```python
import numpy as np;
data=np.loadtxt('data.csv',delimiter=",");
alpha=0.0001;
num_iterations=100;
m,c=grad(data,alpha,num_iterations);
print(m,c);
```
